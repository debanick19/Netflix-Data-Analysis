```python
# Importing Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading CSV File
df = pd.read_csv("netflix_titles.csv")
```

```python
df. sample(10)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python


# Fill missing values with the mode of each respective column
for col in ['country', 'date_added', 'rating', 'duration']:
    df[col] = df[col].fillna(df[col].mode()[0])  # Mode returns a Series, so we take the first mode value

# Verify if null values are removed
print(df.isnull().sum())


```

```python
# Fill missing values in 'director' and 'cast' with "Not Mentioned"
df['director'] = df['director'].fillna("Not Mentioned")
df['cast'] = df['cast'].fillna("Not Mentioned")

# Verify if null values are removed
print(df.isnull().sum())
```

```python
df["rating"].unique()
```

```python
# Define the list of incorrect values to remove
invalid_ratings = ['74 min', '84 min', '66 min']

 #Remove invalid values by replacing them with the mode (most frequent rating)
mode_rating = df['rating'].mode()[0]  # Get the most common rating
df['rating'] = df['rating'].replace(invalid_ratings, mode_rating)

# Verify if the incorrect values are removed
print(df['rating'].unique())


```

```python
# Drop the 'show_id' column
df.drop(columns=['show_id'], inplace=True)
```

```python
df.info()
```

```python
# Rename the columns
df.rename(columns={'year_added': 'Year', 'month_added': 'Month', 'day_added': 'Day'}, inplace=True)
```

```python
df["genre"] = df["listed_in"].str.split(",").str[0]
```

```python
df["genre"].value_counts().head(5)
```

```python
# To find the lead actor from the cast
```

```python
df = df[df["cast"]!="Not Mentioned"]
```

```python
df["lead_actor"] = df["cast"].str.split(",").str[0]
```

```python
df["lead_actor"].value_counts().head(5)
```

```python
# Dropping listed_in and cast column
df.drop(["listed_in","cast"], axis=1, inplace=True)
```
